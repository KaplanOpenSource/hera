# Overview

The 'Hera' package aims to standardize data storage, analysis and presentation. <br>

The library manages data by providing interface to store pieces of data (usually in the form of a file on the disk), 
and their metadata (i.e description of the data). Then, the library provides tools that help with loading, analyzing, and presenting specific data types.

## Hera package structure 

#### Three-tier architecture

The [three-tier architecture](https://en.wikipedia.org/wiki/Multitier_architecture#Three-tier_architecture) divides an application into three layers. That is, 



<table style="font-size: 16px;">
    <tr>
        <td><b>Layer name</b></td>
        <td><b>Description</b></td>
    </tr>
    <tr>
        <td>Data layer</td>
        <td>Manages loading and parsing data</td>
    </tr>
    <tr>
        <td>Analysis layer</td>
        <td>Manages the analysis and computational functions (called logic layer)</td>
    </tr>
    <tr>
        <td>Presentation layer</td>
        <td>Manages the graphs and other types of presentation</td>
    </tr>    
</table>

By isolating each layer, developers can work more efficiently and make updates with minimal impact on other parts of the application.

##### The hera package has the following parts:
 - [Data layer](datalayer/DataLayer.ipynb): Interface to store the metadata (data that describes the data itself)
               of each piece of data as a document in the database

 - [Toolkit](Toolkit.ipynb): Interfaces to manage types of data.

 - [Utils](utils/UtilMain.ipynb): General utils and helpers 

# Defnitions 
## Project 

Each data item is stored in the user database in a **project**. 

The project name can be determined when the data item is added to the database, either through the [low level interface](datalayer/lowlevel.ipynb),
or through the [Project class interface](datalayer/Project.ipynb). 

Alternatively, create a caseConfiguration.json file can provide a default project name for all the 
work performed in that directory. Creating a project is performed with the [hera-project interface](datalayer/Project.ipynb). 

Each user has a default project that holds the [repositories](#Repository) that he loaded. 
When a project is created using the [hera-project interface](datalayer/Project.ipynb), all the datasources 
in all the repositories are loaded into the new project.  

## Toolkit 

A collection of classes (with methods) aimed to handle a single type of data. 
Toolkits have a 3-layer structure (datalayer, analysis, and presentation layer), as detailed in the 
[Toolkit](Toolkit.ipynb) documentation. 


Toolkits are initialized using the [toolkitHome interface](Toolkit.ipynb). 

## Datasource 

Some toolkits can access to external data sources. For example the GIS-topography toolkit can access the 
external file that holds the topography of a region. 

Datasources are managed using the [hera-project CLI or through code](datalayer/Project.ipynb). 

<a href="Repository"></a>
## Repository 

A repository is a collection of datasource that will be loaded when the project is created. 

The repository is managed using the hera-project CLI. 

A repository is stored as a [JSON file](datalayer/Project.ipynb) for details in the structure(TBD).
