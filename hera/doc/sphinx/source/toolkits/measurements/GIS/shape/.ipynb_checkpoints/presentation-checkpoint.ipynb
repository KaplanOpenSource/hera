# Presentation

Saved shapes may be plotted using the shapes tool.
For example, we will plot the saved point added to the database in the previous section.

```python
from hera import toolkitHome

projectName = "documentation"
toolkitName = "GIS_Shapes"
shapes  = toolkitHome.getToolkit(projectName=projectName,toolkitName=toolkitName)

location = "KatsrinCenter"
shapes.presentation.plot(regionNameOrData=location)
```

A desired ax in for the plot and **kwargs for the plot may be delivered as well:

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
shapes.presentation.plot(regionNameOrData=location,ax=ax,color="red",marker="*")
```
