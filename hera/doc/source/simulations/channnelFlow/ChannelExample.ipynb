# Channel flow calculatations. (Using Schlichting text book). 

First, we calculate the required velocity for the required U_star. 

The Ra of different surfaces was taken form Kim et al. 2013 (Functional level...)


```python
from hera.simulations.channelFlow import meanFlow
from unum.units import *
import numpy 
import matplotlib.pyplot as plt 
```

```python
channelHeight = 1.2*mm
nu = 2*1e-5*m**2/s
```

## Smooth vinyl tile 

```python
Ra = 1.5*um
calculator = meanFlow.channelFlow(Ra=Ra,nu=nu,channelHeight=channelHeight)
ustarRange = numpy.arange(1,4.25,0.1)
Um = [calculator.get_Um_from_Ustar(x*m/s).asNumber(m/s) for x in ustarRange]

plt.plot(ustarRange,Um)
plt.plot(ustarRange,10*ustarRange)
```

## Surface with low roughness

```python
Ra = 13*um
calculator = meanFlow.channelFlow(Ra=Ra,nu=nu,channelHeight=channelHeight)
ustarRange = numpy.arange(1,4.25,0.1)
Um = [calculator.get_Um_from_Ustar(x*m/s).asNumber(m/s) for x in ustarRange]

plt.plot(ustarRange,Um)
plt.plot(ustarRange,10*ustarRange)
```
