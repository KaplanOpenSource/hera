# Presentation

Saved images may be plotted using the raster tool.
For example, we will plot the saved image added to the database in the previous section.

```python
from hera import toolkitHome

projectName = "documentation"
toolkitName = toolkitHome.GIS_RASTER
raster  = toolkitHome.getToolkit(projectName=projectName,toolkitName=toolkitName)

location = "Katsrin"
raster.presentation.plot(imageNameOrData=location)
```

A desired ax in for the plot and **kwargs for the plot may be delivered as well:

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
raster.presentation.plot(imageNameOrData=location,ax=ax)
```
