# Data Manipulations

Here we will present some built in manipulations one can perform on geopandas DataFrames.

```python
from hera import GIS
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

manipulator = GIS.dataManipulations()
datalayer = GIS.GIS_datalayer(projectName = "Example", FilesDirectory = "/home/ofir/tmp/BNTL-HAIFA")
```

## Intersection with Polygons

In order to decrease the spacial boundaries of a dataframe to the boundaries of a polygon,
a built in function may be used.
For example, a part of Haifa was saved as a polygon in the dataframe.
The polygon and a dataframe that contains it can be loaded using the GIS datalayer.

```python
polygon = datalayer.getGeometry("PartOfHaifa")
dataframe = datalayer.getGISData(Geometry="PartOfHaifa")[0]
x,y = polygon.exterior.xy

fig, ax = plt.subplots()

ax = dataframe.plot()
ax.plot(x,y, color="black")

plt.show()
```

These shapely polygon and geopandas dataframe can be given as parameters to a function that returns a new dataframe, with the boundaries of the polygon.  

```python
new_dataframe = manipulator.PolygonDataFrameIntersection(dataframe=dataframe, polygon=polygon)

new_dataframe.plot()
plt.show()
```
