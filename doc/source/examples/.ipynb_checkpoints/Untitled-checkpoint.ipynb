# Getting data
This notebook shows how to get the data with the datalayer of pyhera

```python
from pyhera import datalayer
```

After importing the datalayer, you want to search for documents that fits your data properties.
Below we see an example of getting the document of the experimental data of Haifa campaign.

```python
query = {'name':'Haifa'}
docList = datalayer.Experimental.getDocumentsByQuery(query)
haifaDoc = docList[0]
```

Now we want to get the data of specific station, instrument and height, between 2 dates.

```python
station = 'Check_Post'
instrument = 'Sonic'
height = 9
startTime = '01-04-2015'
endTime = '01-05-2015'
data = haifaDoc.getData(station=station,
                        instrument=instrument,
                        height=height,
                        startTime=startTime,
                        endTime=endTime
                       )
```

Now we got the data as dask dataframe.
To get the data as pandas dataframe we need to use the 'compute' function.

```python
data = data.compute()
print(data)
```
